```python
from pathlib import Path
import re
from re import match
import pandas as pd
import spacy
```

The directory `data` contains a subdirectory named `wiki` with 475 random articles from Wikipedia, which are stored as UTF-8 encoded plain text files.

Store these string objects into a list named `wiki_articles`.

```python
corpus_dir = Path('data/wiki')
files = list(corpus_dir.glob('*.txt'))
wiki_articles = []
for file in files:
    text = file.read_text(encoding='utf-8')
    wiki_articles.append(text)
```

```python
wiki_articles[1]
```

Define a regular expression that removes all headers and subheaders from the articles. Store this regular expression under a variable named `pattern`.

```python
pattern = re.compile(r'(\=)\1{1,}')
```

Apply the regular expression under `pattern` to each article (string object) in the list `wiki_articles`. Store each processed article into a new list named `clean_articles`.

```python
clean_articles = []
for article in wiki_articles:
    clean_article = re.sub(pattern, '', article)
    clean_articles.append(clean_article)
```

Extract these headers from the articles, and store them into a dictionary named `titles`.
Each Wikipedia article begins with its title, which is separated from the body text by two line changes (\n)
Use the filename as the key (e.g. `42.txt`) and the article header as the value (e.g. `Rebecca Malope`).

```python
titles = {}
for p in list(Path('data/wiki').glob('*.txt')):
    key = p.name
    with open(file=p, mode="r", encoding="UTF-8") as f:
        text = f.read()
        contents_split = text.splitlines()
        value = contents_split[0]
        #print(key)
    titles.update({key : value})
```
